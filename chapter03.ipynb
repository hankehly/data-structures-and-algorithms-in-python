```python
import math

import matplotlib
import matplotlib.pyplot as plt
```

# Chapter 3 Exercises

## Reinforcement

### R-3.1

```python
xticks = [2 ** i for i in range(6)]

functions = [
    lambda n: 8 * n,
    lambda n: 4 * n * math.log2(n),
    lambda n: 2 * n ** 2,
    lambda n: n ** 3,
    lambda n: 2 ** n
]

labels = [
    "$8n$",
    "$4n\logn$",
    "$2n^2$",
    "$n^3$",
    "$2^n$"
]

ys = []
for f in functions:
    y = list(map(f, xticks))
    ys.append(y)

fig, ax = plt.subplots()

for line, label in zip(ys, labels):
    ax.plot(line, label=label)

ax.legend()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_title("R-3.1")
fig.tight_layout()
plt.show()
```

### R-3.2

```python
def A(n):
    # returns number of operations for some function A with input size n
    return 8 * n * math.log2(n)

def B(n):
    # returns number of operations for some function B with input size n
    return 2 * (n ** 2)

a_v, b_v = [], []
for i in range(1, 25):
    a_v.append(A(i))
    b_v.append(B(i))

fig, ax = plt.subplots()

ax.plot(a_v, label="A")
ax.plot(b_v, label="B")
ax.legend()

plt.show()
```
